# Data Analysis and Binning

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
mc_data = np.loadtxt('mc_data.txt')
```

```python
def get_error(mc_data):
    
    N_bins = mc_data.size
    Δ = np.std(mc_data)/np.sqrt(N_bins)
    start_bin = N_bins % 2
    binned_mc_data = 0.5*(mc_data[start_bin::2]+mc_data[start_bin+1::2])
   
    return Δ,binned_mc_data
```

```python
# number of possible binning levels
num_levels = np.int(np.log2(mc_data.size/4))+1

Δ = []
binned_mc_data = mc_data

for n in range(num_levels):
    Δₙ,binned_mc_data = get_error(binned_mc_data)
    Δ.append(Δₙ)
```

```python
def monte_carlo(N):
    '''Compute pi via Monte Carlo.'''

    hits = np.zeros(N)
    Nhits = 0
    δ = 0.3
    x,y = 1,1
    
    for n in range(N):
        Δx = δ*(-1.0 + 2*np.random.rand())
        Δy = δ*(-1.0 + 2*np.random.rand())
        if np.abs(x+Δx) < 1 and np.abs(y+Δy) < 1:
            x += Δx
            y += Δy

        if x**2 + y**2 < 1:
            hits[n] = 1
            Nhits += 1

    mcpi = 4.0*Nhits / (1.0*N)
    return mcpi,hits
```
